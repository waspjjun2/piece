```python
## N-gram (  )
def ngram_eojeol(s, N=2, opt=False):
    token = s.split()
    ngram = list()
    for i in range(len(token)-(N-1)):
        ngram.append(''.join(token[i:i+N])
                     if opt else tuple(token[i:i+N]))
    return ngram
```

```python
## N-gram (  )
def ngram_umjeol(s, N=2, opt=False):
    ngram = list()
    for i in range(len(s)-(N-1)):
        ngram.append(''.join(s[i:i+N])
                     if opt else tuple(s[i:i+N]))
    return ngram
    
```

```python
##   key(prior)  
def find_key(key, base):
    return list(filter(lambda keys:keys[:len(key)] == key,
                  base.keys()))

#  
def find_prob(key, freq, base):
    return [(_, base[_], base[_]/freq if freq > 0 else 0)
            for _ in find_key(key , base)]

#  model  find_max  
#   
def find_max(key, freq, base):
    return sorted(find_prob(key, freq, base),
                 key=lambda _: _[-1],
                 reverse=True)[0]

# 
def ngram_print(key, n, base1, base2):
    for _ in range(n):
        nextkey = find_max(key, base1[key], base2)
        print(key)
        print(nextkey)

        key = nextkey[0][-len(key):]
```

```python
# BPE

## input  
def convert_data(data): # {'': }
    newdata = dict()
    for k,v in data.items():
        newdata[' '.join(list(k) + ['</w>'])] = v
    return newdata

def find_pair(data): #  =>  
    pair = defaultdict(lambda:0)
    for k, v in data.items(): # l o w </w> : 5
        tokens = k.split() # [l, o , w, </w>]
        for i in range(len(tokens)-1): #    -1 
                                        #ex. (l, o):5, (o, w):5
            pair[tuple(tokens[i:i+2])] += v
    return pair


# max pattern -> 
import re

def merge_pair(data, key):
    newdata = dict()
    for k, v in data.items():
        newkey = re.sub(' '.join(key), ''.join(key), k) # 'e s'
        newdata[newkey] = v
    return newdata

# re.sub 
```

```python
# NLP -> 언어 패턴 분석 
# (Language Model) - statistical LM
# s = '대한민국의 주권은 국민으로부터 나온다.'
# t1 = 대한민국의, 100
# t2 = 주권은, 10 
# t3 = 국민으로부터, 200
# t4 = 나온다, 500
# t5 = ., 1000
# N = 1000
# p(t5)는 얼마나 될까..?
# p(t5) = freq(t5)/N = 1.0
# p(t1) = freq(t1)/N = 0.1
# p(S) = p(t1,t2,t3,t4,t5) = p(t2,t3,t4,t5|t1)p(t1)
#         결합확률                CP        P
#     (JP - 못구함... 문법을 지키는 경우 문장의 다양한 형태가 나올 수 x
#      -> 조건부확률 적용!)
#         p(t3,t4,t5|t1,t2)p(t2|t1)p(t1)
#         p(t5|t1,t2,t3,t4)p(t1,t2,t3,t4)
#                          p(t4|t1,t2,t3)p(t1,t2,t3)
#                                        p(t3|t1,t2)p(t1,t2)
#                                                   p(t2|t1)p(t1)
                    
# S = '대한민국의 주권은 국민에게 나온다.'
# t6 = '국민에게'           
# P(S) = p(t1,t2,t3,t6,t5)
# -> 이런건 우리가 가진 데이터에 없는데? 0으로 반환하는 문제
# -> 1st Markov Assumption(Chain)
# P(S) = p(t5|(t1,t2,t3),t4)p(t4|(t1,t2),t3)p(t3|(t1,)t2)p(t2|t1)p(t1)
#      = p(t5|t4)p(t4|t3)p(t3|t2)p(t2|t1)p(t1)
    
# ## N-gram; Bi-gram
# p(t5|t4) = p(t4,t5)/p(t4)
#            =>freq(t4,t5)(/N)         => freq(t4)(/N)
#         = freq(t4,t5)/ freq(t4)

# p(?|t4) = sum(freq(?,t4)/freq(t4))
#         = (t4,t1), (t4,t2), (t4, t3), (t4, t5), (t4, t6)
#             -> freq 이용

### n-gram의 문제점?? (1)그래서 실제로 1st MC가 맞아들어가나
#                    (2)0이 나올 경우 nextkey 부분에서 max 찾다가 오류

```

```python
## N-gram ( 어절 )
def ngram_eojeol(s, N=2, opt=False):
    token = s.split()
    ngram = list()
    for i in range(len(token)-(N-1)):
        ngram.append(''.join(token[i:i+N])
                     if opt else tuple(token[i:i+N]))
    return ngram
```

```python
## N-gram ( 음절 )
def ngram_umjeol(s, N=2, opt=False):
    ngram = list()
    for i in range(len(s)-(N-1)):
        ngram.append(''.join(s[i:i+N])
                     if opt else tuple(s[i:i+N]))
    return ngram
    
```

```python
## 어떤 그램이든 key(prior) 찾아오는 
def find_key(key, base):
    return list(filter(lambda keys:keys[:len(key)] == key,
                  base.keys()))

# 확률값 계산
def find_prob(key, freq, base):
    return [(_, base[_], base[_]/freq if freq > 0 else 0)
            for _ in find_key(key , base)]

# 띄어쓰기 model 만드려면 find_max에 무언가를 추가
# 가장 높은 확률값
def find_max(key, freq, base):
    return sorted(find_prob(key, freq, base),
                 key=lambda _: _[-1],
                 reverse=True)[0]

# 출력
def ngram_print(key, n, base1, base2):
    for _ in range(n):
        nextkey = find_max(key, base1[key], base2)
        print(key)
        print(nextkey)

        key = nextkey[0][-len(key):]
```

```python
# BPE

## input 형식 만들기
def convert_data(data): # {'문자열': 빈도}
    newdata = dict()
    for k,v in data.items():
        newdata[' '.join(list(k) + ['</w>'])] = v
    return newdata

def find_pair(data): # 문자열 => 쌍으로 변환
    pair = defaultdict(lambda:0)
    for k, v in data.items(): # l o w </w> : 5
        tokens = k.split() # [l, o , w, </w>]
        for i in range(len(tokens)-1): # 쌍으로 변환 위해 -1 
                                        #ex. (l, o):5, (o, w):5
            pair[tuple(tokens[i:i+2])] += v
    return pair


# max pattern -> 합치기
import re

def merge_pair(data, key):
    newdata = dict()
    for k, v in data.items():
        newkey = re.sub(' '.join(key), ''.join(key), k) # 'e s'
        newdata[newkey] = v
    return newdata

# re.sub 찾아보기
```
